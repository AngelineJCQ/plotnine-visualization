Here is the dataset. I've added a column to make the `am` variable more human readable. You may want to do the same for a few other columns too if you plan to put them in your charts:

```python
from plotnine.data import mtcars
from plotnine import ggplot, aes, geom_point, labs, theme_bw, facet_wrap, geom_smooth

# give transmission_type a human readable name rather than 0 or 1
mtcars['transmission_type'] = mtcars['am']\
    .apply(lambda x: {0: 'automatic', 
                      1: 'manual'}[x])

# Extract `brand` from name
mtcars['brand'] = mtcars['name'].str.split(expand=True)[0]

# Display the dataset
mtcars
```

Let's start doing some exploratory data visualization with plotnine! Take a look at the example below:

```python
display(
    ggplot(mtcars,                   # data
        aes('mpg', 'disp'))          # aesthetics
        + geom_point()               # geometry
        + theme_bw()
        + labs(
            title = "Cars with big engines tend to be less efficient",
            y = "engine displacement (cu.in.)",
            x = "fuel efficiency (mpg)"
        )
)   
```

**Summary:**

According to [Soren Heitmann](https://rstudio-pubs-static.s3.amazonaws.com/61800_faea93548c6b49cc91cd0c5ef5059894.html), the engine displacement of a car is a good proxy for the total amount of power the engine can generate. As you can see in the chart above, among the cars that Motor Trend looked at, those with that generated more power tended to be less fuel efficient.

**Reflection**

What I can conclude based on this chart:
- Among the cars that motortrend analyzied, the cars with bigger engines have lower fuel efficiency
- Bigger engines are correlated with lower fuel efficiency

What I can't conclude based on this chart:
- Bigger engines are less efficient. 
    - In order to come to this conclusion, I would also need to 👉_control other factors, for instance, the same car type, number of cylinders, foward gears, etc.._______________

```python
# 👉Chart 1
# Insert code for a plotnine chart below
display(
    ggplot(mtcars,                   # data
        aes('mpg', 'cyl'))          # aesthetics
        + geom_point(alpha=.7)               # geometry
        + theme_bw()
        + facet_wrap('transmission_type')
        + labs(
            title = "Cars with more cylingders tend to be less efficient",
            y = "number of cylinders",
            x = "fuel efficiency (mpg)"
        )
)  
```

**👉Chart 1 Summary**

According to [Soren Heitmann](https://rstudio-pubs-static.s3.amazonaws.com/61800_faea93548c6b49cc91cd0c5ef5059894.html), as well as indicated in charts, the dataset includes vehicles with 4,6,8 cylinder engines. Among the cars that Motor Trend looked at, cars with more cylingders tend to be less efficient. It's more commom for automatic cars to have eight cylinders and manual cars to have only four cylinders.

**👉Reflection**

What I can conclude based on this chart:
- Among the cars that Motor Trend looked at, cars with more cylingders tend to be less efficient.
- In general, there is higher possibility for automatic cars to be equipped with more cylinders.
- The range of fuel efficiency is much bigger among cars with four cylingders than eight.
    
What I can't conclude based on this chart:
- Automatic cars have more cylinders than manual cars.
- More cylingders lead to less fuel efficiency.


```python
# 👉Chart 2
# Insert code for a plotnine chart below
from plotnine import geom_histogram
display(
    ggplot(mtcars, aes(x='wt', fill='transmission_type')) 
    + geom_histogram(binwidth=.3, alpha=.5, position="identity")
    + labs(
        title = "Automatic cars are heavier than manual ones in general",
        y = "Number of cars",
        x = "Weight (lb/1000)"
        )
)
```

**👉Chart 2 Summary**

As you can see from the chart, the automatic car is usually heavier than the manual one.

**👉Reflection**

What I can conclude based on this chart:
- Among the cars that motortrend analyzied, the overall weight of automatic car is heavier than the manual kind.
- Most of automatic cars are above 3000 lbs, while there are only a few manual types weight exceeding 3000.
    
What I can't conclude based on this chart:
- The automatic car is heavier than the manual one.


```python
# 👉Chart 3
# Insert code for a plotnine chart below
from plotnine import stat_smooth
display(
    ggplot(mtcars,                   # data
        aes('disp','wt',color='factor(vs)'))          # aesthetics
        + geom_point()               # geometry
        + theme_bw()
        + labs(
            title = "There is a correlation between the car weight and the engine power",
            y = 'weight',
            x = 'engine displacement (cu.in.)'
        )
        + stat_smooth(method='lm') 
)  
```

**👉Chart 3 Summary**

According to [Soren Heitmann](https://rstudio-pubs-static.s3.amazonaws.com/61800_faea93548c6b49cc91cd0c5ef5059894.html), the engine displacement of a car is a good proxy for the total amount of power the engine can generate. Among the cars that Motor Trend looked at, there is an association between the weight and the engine power. As indicated by color, those heavy and powerful cars always carry straight line engine cylinder.


**👉Reflection**

What I can conclude based on this chart:
- Among the cars that Motor Trend looked at, heavier cars usally carry more powerful engines.
- Those heavy and powerful cars always carry straight line engine cylinder instead of V-shape.
    
What I can't conclude based on this chart:
- Cars with more powerful engines are heavier.
- Heavier cars are more powerful.

^^ They are more causal relations.

```python
# 👉Chart 4
# Insert code for a plotnine chart below
display(
    ggplot(mtcars,                   # data
        aes('qsec','mpg',fill='transmission_type'))          # aesthetics
        + geom_point()               # geometry
        + theme_bw()
        + labs(
            title = "The manual cars tend to be more fuel-efficient than automatic",
            x = 'acceleration (1/4 mile second)',
            y = 'fuel efficiency (mpg)'
        )
        + stat_smooth(method='lm') 
)  

```

**👉Chart 4 Summary**

According to [Soren Heitmann](https://rstudio-pubs-static.s3.amazonaws.com/61800_faea93548c6b49cc91cd0c5ef5059894.html), qsec is an indicator of acceleration speed and mpg shows the fuel efficiency of a car. Among the cars that Motor Trend looked at, manual cars are faster and have higher fuel efficiency overall.

**👉Reflection**

What I can conclude based on this chart:
- Most cars in the dataset can run 10-20 miles per gallon.
- Most automatic cars we look can can run 1/4 mile in 16-20 seconds.
- When reaching the same acceleration speed, the manual cars we look can drive further distance per gallon overall, which means they are more fuel-efficient than automatic ones.
    
What I can't conclude based on this chart:
- Manual cars are faster and fuel-efficient than automatic ones under any condition.
