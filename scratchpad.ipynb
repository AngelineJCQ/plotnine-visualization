```python
from plotnine.data import mtcars

# Display the dataset
mtcars.head()
```

Make your plotnine charts below! Don't forget to import each component

```python
# disp = engine displayment
# mpg = miles per gallon
# am = auto/mannual
```

```python
from plotnine import ggplot, aes, geom_point, facet_wrap, stat_smooth
```

```python
(ggplot(mtcars, aes(x='disp',y='mpg'))+
    geom_point())
```

```python
(ggplot(mtcars)+
    geom_point(aes(x='disp',y='mpg')))
```

```python
(ggplot(mtcars, aes(x='disp',y='mpg'))+
    geom_point(shape='*',size=6))
```

```python
(ggplot(mtcars, aes(x='disp',y='mpg'))+
    geom_point(shape='*',size=6)+
    facet_wrap('am')
)
```

```python
(ggplot(mtcars, aes(x='disp',y='mpg'))+
    geom_point(shape='*',size=6)+
    facet_wrap('am')+
    stat_smooth(method='mavg')
)
# lm = linear model
# mavg = moving average
```

```python
from plotnine import coord_cartesian, theme_bw, coord_flip
```

```python
(ggplot(mtcars, aes(x="disp", y="mpg"))+
    geom_point() +
    facet_wrap('am') +
    stat_smooth(method='lm')+
    coord_cartesian(xlim = [0, 500], 
                    ylim = [0, 50]) +
    coord_flip() #change the axis x and y
)
```
